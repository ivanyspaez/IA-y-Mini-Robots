```python
class Pila:
    def __init__(self):
         self.items = []

    def pruebaPila(self):
         return self.items == []

    def insertaPila(self, item):
         self.items.append(item)

    def eliminaPila(self):
         return self.items.pop()

    def accedePila(self):
         return self.items[len(self.items)-1]

    def tamanoPila(self):
         return len(self.items)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Voronoi
import matplotlib.image as mpimg

# 1. Coordenadas de ejemplo (muestra; debes reemplazar con datos reales)
pharm = np.array([[4.87786, -74.03735], [4.8758, -74.0385], [4.8730, -74.0440], [4.8630, -74.0590]])
health = np.array([[4.85637, -74.06263], [4.8625, -74.0586], [4.8700, -74.0450]])
schools = np.array([[4.8890, -74.0390], [4.8740, -74.0445], [4.8800, -74.0350], [4.8635, -74.0595]])

# 2. Leer mapa base
img = mpimg.imread('mapa_chia.png')  # descarga la imagen mostrada arriba

# 3. Función para Voronoi sobre mapa
def plot_voronoi_on_map(points, title, color):
    pts = np.column_stack((points[:,1], points[:,0]))
    vor = Voronoi(pts)
    fig, ax = plt.subplots(figsize=(8,8))
    ax.imshow(img, extent=[-74.07, -74.01, 4.84, 4.92], aspect='equal')  # ajustar según imagen
    # dibujar regiones
    for region in vor.regions:
        if not -1 in region and region:
            poly = [vor.vertices[i] for i in region]
            ax.fill(*zip(*poly), alpha=0.2, color=color)
    ax.plot(pts[:,0], pts[:,1], 'o', color=color)
    ax.set_title(title)
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')
    plt.show()

# 4. Generar diagramas
plot_voronoi_on_map(pharm, "Voronoi: Droguerías", 'blue')
plot_voronoi_on_map(health, "Voronoi: Centros de Salud", 'green')
plot_voronoi_on_map(schools, "Voronoi: Colegios", 'red')

# 5. Combinado
all_pts = np.vstack([pharm, health, schools])
plot_voronoi_on_map(all_pts, "Voronoi Combinado - Servicios", 'magenta')
```
